```python
import pandas as pd
from astropy import units as u
from astropy.coordinates import SkyCoord
from astropy.table import Table
```

```python
# csv table from copy element from inspect the web source code and converting to csv using https://www.convertcsv.com/html-table-to-csv.htm

df = pd.read_csv('./data/updates/UltraCompCAT.csv')

df['ra'] = SkyCoord(df['RA'], df['Dec'], unit=(u.hourangle, u.deg), frame='icrs').ra.deg
df['dec'] = SkyCoord(df['RA'], df['Dec'], unit=(u.hourangle, u.deg), frame='icrs').dec.deg



# 38: should match to 2CXO J171236.7-373841, ra=258.15322 dec=-37.644746
df.loc[df['ID']==38, 'ra'] = 258.15322
df.loc[df['ID']==38, 'dec'] = -37.644746 

df[['ID','RA','Dec','ra','dec']]
```

```python
CSCv2 = Table.read('/Users/huiyang/Research/GitHub/MUWCLASS_CSCv2/codes/buildTD/data/CSCv2.vot').to_pandas()
CSCv2['name'] = '2CXO ' + CSCv2['_2CXO']
```

```python
CSCv2[CSCv2['name']=='2CXO J171236.7-373841']
```

```python
c = SkyCoord(ra=df['ra']*u.degree, dec=df['dec']*u.degree)
catalog = SkyCoord(ra=CSCv2['RAICRS']*u.degree, dec=CSCv2['DEICRS']*u.degree)
idx, d2d, d3d = c.match_to_catalog_sky(catalog,nthneighbor=1)
```

```python
df['CXO'] = CSCv2.loc[idx, '_2CXO'].values
df['sep'] = d2d.arcsec
df['CXO_name'] = '2CXO '+df['CXO']

df.loc[df['sep']<5, ['ID','RA','Dec', 'CXO', 'sep']]

df_sel = df[df['sep']<5].reset_index(drop=True)

```
