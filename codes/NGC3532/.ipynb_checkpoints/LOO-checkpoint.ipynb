```python
import pandas as pd
import numpy as np
import os

# import time
import multiprocessing as mp

from test_library import loo_prepare, loo_train_and_classify, loo_save_res, plot_confusion_matrix, plot_CM_withSTD
from collections import Counter
from sklearn.metrics import confusion_matrix

import matplotlib.pyplot as plt

pd.options.mode.chained_assignment = None
```

```python
dir_tool = 'data/tools'

Xcat = 'CSC'
distance = 'rgeo_lum'
# distance = 'nodist'
red_switch = True
# seed for generating all randomness, set to none
random_state = None
Uncer_flag = True

dir_out = './LOO_results/' + distance
dir_plot = 'C:/Users/Steven/Dropbox/Research/MUWCLASS/' + distance

n_estimators = 1000 # 1000
```

```python
os.system(f'mkdir -p {dir_out}')
os.makedirs(dir_plot)
```

```python
df = pd.read_csv(f'./data/TD/CSC_TD_v5_MW_remove.csv') 
df = df[df.remove_code==0].reset_index(drop=True)

df = df.rename(columns = {
    'W1mag_comb':'W1mag','W2mag_comb':'W2mag','W3mag_allwise':'W3mag','W4mag_allwise':'W4mag',
    'e_W1mag_comb':'e_W1mag','e_W2mag_comb':'e_W2mag','e_W3mag_allwise':'e_W3mag','e_W4mag_allwise':'e_W4mag',
    'kp_prob_b_max':'var_intra_prob'})

# CSC
for band in ['s','m','h']:
    df['Fcsc_' + band], df['e_Fcsc_' + band] = df['flux_aper90_ave_' + band], df['e_flux_aper90_ave_' + band]

df['e_rgeo'] = (df['B_rgeo'] - df['b_rgeo'])/2
df['e_rpgeo'] = (df['B_rpgeo'] - df['b_rpgeo'])/2

'''
# XMM 
for band in ['2','3','4','5','8']:
    df['Fxmm_' + band], df['e_Fxmm_' + band] = df['SC_EP_'+band+'_FLUX'], df['e_SC_EP_'+band+'_FLUX']
'''

print(len(df))
# df = df.sample(frac=0.3).drop_duplicates(subset=['name']).sort_index().reset_index(drop=True) # ?
print(Counter(df['Class']))

source_names = df.name
```

```python
# tbabs from xpsec Wilms, Allen & McCray (2000)
# return in unit of 1.e-24 cm^2
tbabs_ene, tbabs_cross = np.loadtxt(f'{dir_tool}/tbabs.data', skiprows=0)
tbabs_cross *= 100
```

```python

%time arr = [loo_prepare(i, df, red_switch, Xcat, distance, Uncer_flag, random_state, tbabs_ene, tbabs_cross) for i in range(len(df))]

'''
num_cores = mp.cpu_count()
pool = mp.Pool(int(num_cores))

%time arr = [pool.apply(loo_prepare, args=(i,df, red_switch, Xcat, distance, Uncer_flag, random_state, tbabs_ene, tbabs_cross)) for i in range(len(df))]

#CPU times: user 5.21 s, sys: 2.91 s, total: 8.12 s
#Wall time: 6min 10s

'''
```

```python
opts = {'bootstrap': True, 'class_weight': None, 'criterion': 'gini',
        'max_depth': None, 'max_features': 5, 'max_leaf_nodes': None,
        'min_samples_leaf': 1, 'min_samples_split': 2,
        'min_weight_fraction_leaf': 0.0, 'n_jobs': 1,#-1,
        'random_state': random_state, 'verbose': 0, 'warm_start': False,
        'n_estimators': n_estimators}
```

```python
%%time

arr = [_ for _ in zip(arr, [opts]*len(arr))]

if __name__ == '__main__':
    
    with mp.Pool() as pool:
        res = pool.map(loo_train_and_classify, arr) 
```

```python
%time loo_save_res(res, dir_out)
```

# Plotting

```python
from bokeh.io import output_notebook, show
import colorcet as cc
from bokeh.layouts import layout
output_notebook()
```

```python

df_classes = pd.DataFrame()
for i in range(100):
    df_classes = df_classes.append(pd.read_csv(f'{dir_out}/{str(i)}/classes.csv'))

p = plot_confusion_matrix(df_classes, 
                          title='Normalized confusion matrix (%)',
                          normalize=True,                          
                          pallete=[cc.blues, cc.kbc[::-1], cc.fire[::-1]][-1],
                          fill_alpha=0.6,
                          width=600, 
                          height=600,
                          plot_zeroes=True
                         )

p70 = plot_confusion_matrix(df_classes[df_classes['Class_prob']>=0.7], 
                          title='Normalized confusion matrix >= 70%',
                          normalize=True,                          
                          pallete=[cc.blues, cc.kbc[::-1], cc.fire[::-1]][-1],
                          fill_alpha=0.6,
                          width=600, 
                          height=600,
                          plot_zeroes=True
                         )


show(layout([[p, p70]]))
```

```python
df_all = pd.DataFrame()
confusions = []
confusions_70 = []

for i in range(0, 100):
    
    data = pd.read_csv(f'{dir_out}/{str(i)}/classes.csv')

    classes = ['AGN',
            'NS',
            'NS_BIN',
            'CV',
            'HM-STAR',
            'HMXB',
            'LM-STAR',
            'LMXB',
            'YSO',
            'true_Class',
            'Class',
            'Class_prob',
            'name']
    data = data[classes]
    
    probs = np.array(data.iloc[:,:9])
    probs_max = np.amax(probs,axis =1)
    where_70 = np.where(probs_max >=0.7)[0]
    if where_70.any():
        name_70 = data.iloc[where_70]['name']
        probs_70 = probs[where_70]
        probs_max_70 = probs_max[where_70]
        y_true_70 = data.iloc[where_70]['true_Class']# .real_class.values[where_70]
        y_pred_70 = data.iloc[where_70]['Class']

        confusions_70.append(confusion_matrix(y_true_70, y_pred_70, labels=data.columns[:9]))

    confusions.append(confusion_matrix(data.true_Class, data.Class, labels=data.columns[:9]))
    df_all = df_all.append(data)

fig = plot_CM_withSTD(np.average(confusions, axis=0),np.std(confusions, axis=0),classes=data.columns[:9], normalize=True)
plt.savefig(f'{dir_plot}/LOO_CSC_Uncer_100ave_CM_{distance}.png')
# plt.close(fig)

fig = plot_CM_withSTD(np.average(confusions_70, axis=0),np.std(confusions_70, axis=0),classes=data.columns[:9], normalize=True, title='Normalized confusion matrix >70%')
plt.savefig(f'{dir_plot}/LOO_CSC_Uncer_100ave_70CM_{distance}.png')
```
